```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("csv_file/young_survey.csv")
df.head()
```

```python
info = df.iloc[:,140:] # 조사 대상의 기본 정보를 알아보기 위해
info.head()
```

```python
info.describe() # Gender, Handedness, Education는 string이기 때문에 통계자료를 표현하지 않음
```

```python
info["Gender"].value_counts()
```

```python
info["Handedness"].value_counts()
```

```python
info["Education"].value_counts()
```

```python
sns.violinplot(data=info, y = "Age")
```

```python
sns.violinplot(data=info, x = "Gender", y = "Age") # 성별에 따른 나이 관계
```

```python
sns.violinplot(data=info, x = "Gender", y = "Age", hue = "Handedness") # 손잡이와 성별에 따른 나이 관계
```

```python
sns.jointplot(data = info, x = "Height", y = "Weight") # jointplot은 두 데이터간의 산점도와 각각의 bar 그래프를 같이 보여준다! 짱짱맨~
```

```python
df = pd.read_csv("csv_file/young_survey.csv")

music = df.iloc[:, :19] # 음악과 관련된 데이터
music.head()
```

# 클러스터 분석
*비슷한 성향을 가진 몇가지 무리에 대해서 분석하는 방법*

```python
interests = df.loc[:,"History":"Pets"]
interests.head()
```

```python
corr = interests.corr()
corr
```

```python
corr["History"].sort_values(ascending = False)
```

```python
sns.clustermap(corr) # clustermap => heatmap과 비슷하지만, 연관성있는 관심사들끼리 묶어놓은 결과도 제공한다!
```
