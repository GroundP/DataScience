# 선 그래프 

*선그래프는 특정 항목의 변화를 나타내고 싶을때 쓰임*

```python
%matplotlib inline

import pandas as pd
```

```python
df = pd.read_csv("csv_file/broadcast.csv", index_col = 0)
df
```

```python
df.plot() # plot 메소드(선 그래프가 디폴트)
```

```python
df.plot(y = "KBS") # y 파라미터는 특정 항목만 보고싶을때
```

```python
df.plot(y = ["KBS", "JTBC"])
```

```python
df[["KBS", "JTBC"]].plot()
```

# 막대그래프

```python
%matplotlib inline

import pandas as pd

df = pd.read_csv("csv_file/sports.csv", index_col = 0)
df
```

```python
df.plot(kind = "bar") # 막대그래프는 kind = "bar"
```

```python
df.plot(kind = "barh") # h는 horizontal의 약자
```

```python
df.plot(kind = "bar", stacked = True) # stacked 매개변수는 쌓아올리는 변수
```

```python
df["Female"].plot(kind = "bar")
```

# 히스토그램
* 특정 항목들의 묶음으로 대략적인 값들의 분포를 나타내고 싶을 때(키값이 많을때)

```python
import pandas as pd

df = pd.read_csv("csv_file/body.csv", index_col = 0)
df
```

```python
df.plot(kind = "hist", y = "Height") # kind = "hist"(histogram의 약자)
```

```python
df.plot(kind = "hist", y = "Height", bins = 20) # bins는 값을 쪼개는 단위를 설정한다.(숫자가 클수록 많이 쪼개겠찌)
```

# 박스 플롯
*어떤 데이터들에 대한 정보를 통계적으로 보여줄 때*

```python
df = pd.read_csv("csv_file/exam.csv")
df.head()
```

```python
df["math score"].describe() # 수학 정보의 통계
```

```python
df.plot(kind = "box", y = "math score") # 박스 플롯 출력
```

```python
# 27~28점 이하는 이상점으로 처리가 되었음
```

```python
df.plot(kind = "box", y = ["math score", "reading score", "writing score"]) # 여러개의 박스 플롯 출력
```

```python
# 상위 25%와 하위 25%가 긴걸로 봐서는 그 부분에 해당하는 학생들의 점수가 분산되어 있음을 알 수 있다
```

# 산점도(Scatter plot)

*특정 항목들의 상관관계를 나타내고 싶을때*

```python
df.plot(kind = "scatter", x="math score", y = "reading score") # 수학 점수와, 읽기 점수간 상관관계를 위한 산점도 출력
```

```python
df.plot(kind = "scatter", x="math score", y = "writing score") # 수학 점수와, 쓰기 점수간 상관관계를 위한 산점도 출력
```

```python
df.plot(kind = "scatter", x="reading score", y = "writing score") # 읽기 점수와, 쓰기 점수간 상관관계를 위한 산점도 출력
```

```python
# 인사이트 : 시험을 잘보는 학생은 대체적으로 점수가 높다(어느 한 과목만 높은 사람은 없었음)
#            읽기 점수와 쓰기 점수는 확실히 상관관계가 있다(산점도가 더욱 일직선이므로)
```
