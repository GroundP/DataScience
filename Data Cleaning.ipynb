# 완결성 갖추기

*결측값이 있는 데이터 Cleaning*

```python
# 결측값이 없도록 한다.
# 결측값이 자주 있다면 원인을 파악해야 한다.
```

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv("csv_file/attendance.csv")
df
```

```python
df.isnull() # 결측값 여부
```

```python
df.isnull().sum() # 결측값의 개수
```

*결측값이 있는 데이터를 어떻게 다뤄야 하나*

```python
# 결측값이 있는 레코드를 삭제하는 방법
df.dropna(inplace = False) # True면 원본 바뀜
```

```python
# 결측값이 있는 Column을 삭제하는 방법
df.dropna(axis = "columns", inplace = False)
```

```python
# 결측값에 데이터 넣기
df.fillna(0) # fillna는 결측값에 값을 넣는 메서드
```

```python
# 결측값에 평균 데이터를 넣어버리기
df.mean()
```

```python
df.fillna(df.mean())
```

```python
# 결측값에 중간값을 넣어버리기
df.fillna(df.median())
```

# 정확성 갖추기

*이상점을 어떻게 다룰 것인가*

```python
df = pd.read_csv("csv_file/beer.csv", index_col = 0)
df.head()
```

```python
df.plot(kind = "box", y = "abv")
```

```python
df["abv"].describe() # 뜬금없는 5.5값
```

```python
q1 = df["abv"].quantile(0.25)
q1
```

```python
q3 = df["abv"].quantile(0.75)
q3
```

```python
iqr = q3 - q1
iqr
```

```python
condition = (df["abv"] < q1 - 1.5*iqr) | (df["abv"] > q3 + 1.5*iqr)
condition
```

```python
df[condition] # 이상점 데이터만 골라보자~
```

```python
# 참이슬 후레시, 보드카가 왜 beer데이터에 들어가있냐 ㅡ.ㅡ 이상점이 잘못된 데이터임
# APA는 5.5%일텐데 그러면 0.055의 값이어야 하는데 5.5로 잘못들어간듯... 
# 예측할 수 있다!
```

```python
df.loc[2250, "abv"] = 0.055     # 제대로된 값으로 변경
df.loc[2250]
```

```python
condition = (df["abv"] < q1 - 1.5*iqr) | (df["abv"] > q3 + 1.5*iqr)
df[condition]
```

```python
df[condition].index
```

```python
df.drop(df[condition].index, inplace = True)
df[condition]
```

```python
df.plot(kind = "box", y = "abv") # 이상점을 제거/수정해서 좀더 정확한 데이터를 갖게 되었다!
```
