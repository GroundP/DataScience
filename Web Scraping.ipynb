# 파싱

*파이썬에서는 Beautiful Soup이라는 툴로 HTML을 파싱한다.*

![BeautifulSoup](img/rGxaSlb.png)

```python
html_code = """
<!DOCTYPE html>
<html>
<head>
    <title>Sample Website</title>
</head>
<body>
<h2>HTML 연습!</h2>

<p>이것은 첫 번째 문단입니다.</p>
<p>이것은 두 번째 문단입니다!</p>

<ul>
    <li>커피</li>
    <li>녹차</li>
    <li>우유</li>
</ul>

<img src='https://i.imgur.com/bY0l0PC.jpg' alt="coffee"/>
<img src='https://i.imgur.com/fvJLWdV.jpg' alt="green-tea"/>
<img src='https://i.imgur.com/rNOIbNt.jpg' alt="milk"/>

</body>

</html>"""
```

```python
from bs4 import BeautifulSoup
# bs4 라이브러리의 BeautifulSoup 모듈 불러옴
```

```python
# BeautifulSoup 타입으로 변환
soup = BeautifulSoup(html_code, 'html.parser')

# type 출력
print(type(soup))
```

```python
# 모든 <li> 태그 선택하기
li_tags = soup.select('li')

# 결과 출력
print(li_tags)
```

```python
type(li_tags[0])
```

```python
# 첫 번째 <li> 태그 출력하기
print(li_tags[0])

# 첫 번째 <li> 태그의 텍스트 출력하기
print(li_tags[0].text)
```

```python
beverage_names = []

# 텍스트 추출해서 리스트에 담기
for li in li_tags:
    beverage_names.append(li.text)

# 결과 출력
print(beverage_names)
```

```python
# 이미지 태그 가져오자
img_tags = soup.select('img')
print(img_tags)
```

```python
print(img_tags[0])
```

```python
# 첫 번째 요소의 "src" 속성 값 가져오기
print(img_tags[0]["src"])
```

```python
img_srcs = []

# 이미지 주소 추출해서 리스트에 담기
for img in img_tags:
    img_srcs.append(img["src"])

# 결과 출력
print(img_srcs)
```

# 필요한 데이터만 가져오기

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.ikea.com/kr/ko/search/?query=desk")
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# 이케아 홈페이지에서 'desk'를 검색한 결과가 몇페이지까지 나오는지 확인해보자

page_list = soup.select(".pagination a")
page_list
```

```python
page_text_list = []
for page in page_list:
    page_text_list.append(page.text)
    
page_text_list[-1] # 마지막 페이지는 12
```

```python
# 만약에 페이지 표시도 없을 땐,
# 일단은 페이지를 가져와서 데이터가 비어 있는지 확인하는 방법도 있다
```

```python
# 페이지의 처음부터 반복해나가면서, 비어있는지를 확인한다

pages = []
page_num = 1

while True:
    response = requests.get("https://www.ikea.com/kr/ko/search/?query=desk&pageNumber=" + str(page_num))
    soup = BeautifulSoup(response.text, 'html.parser')
    
    
    if len(soup.select(".prodName")) > 0: # 비어있지 않은 페이지에 존재하는 prodName클래스로 빈페이지 여부 확인
        pages.append(soup)
        page_num += 1
    else:
        break

print(len(pages))
```

# 웹페이지를 DataFrame으로!

```python
import requests
from bs4 import BeautifulSoup
```

```python
records = []
page_num = 1

while True:
    response = requests.get("https://www.ikea.com/kr/ko/search/?query=desk&pageNumber=" + str(page_num))
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 페이지에 데이터가 있을 경우에만
    if ( len(soup.select('.prodName')) > 0 ):
        names = soup.select('.prodName')
        prices = soup.select('.prodPrice')
        urls = soup.select('.prodImg')
        page_num += 1
        
        
        # 상품의 정보를 하나의 레코드로 만들고, 리스트에 순서대로 추가하기
        for i in range(len(names)):
            record = []
            record.append(names[i].text)
            record.append(prices[i].text.strip())
            record.append("https://www.ikea.com" + urls[i].get('src'))
            records.append(record)
            
    else:
        break
    
# 결과 출력
print(len(records))
print(records)
```

```python
import pandas as pd

# DataFrame 만들기
df = pd.DataFrame(data = records, columns = ["이름", "가격", "이미지 주소"])
df.head()
```
