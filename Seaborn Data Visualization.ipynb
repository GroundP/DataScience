```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("csv_file/body.csv", index_col = 0)
df.head()
```

```python
df["Height"]
```

```python
df["Height"].value_counts()
```

```python
df["Height"].value_counts().sort_index() # index값을 정렬
```

```python
df["Height"].value_counts().sort_index().plot()
```

```python
sns.kdeplot(df["Height"]) # Seaborn 라이브러리의 kdeplot이라는 함수
```

```python
# 확률 밀도 함수의 그래프가 나옴, 선 밑의 면적은 딱 1임
```

```python
sns.kdeplot(df["Height"], bw = 0.5) # bw는 실제 데이터를 얼마나 덜/더 가공하는지에 대한 정도(band width)
```

```python
sns.kdeplot(df["Height"], bw = 0.1) # bw가 너무 크면 부정확한 결과가 나올 수 있고, 너무 작으면 보기 힘듦 -> 적당히!
```

```python
df.plot(kind = "hist", y = "Height", bins = 15)
```

```python
sns.distplot(df["Height"], bins = 15) # 히스토그램 위에 kde곡선이 같이 그려짐(히스토그램과 kde곡선은 비슷한 형태임(데이터의 분포를 나타내기 때문))
```

```python
df.plot(kind = "box", y = "Height")
```

```python
sns.violinplot(y = df["Height"]) # 바이올린플롯(분포 전체를 보여준다~)
```

```python
df.plot(kind = "scatter", x = "Height", y = "Weight")
```

```python
sns.kdeplot(df["Height"], df["Weight"]) 
# 등고선 형태의 그래프, x축에는 Height에 대한 kde, y축에 대한 kde
# 선끼리의 거리가 가까울 수록 가파르고, 멀수록 평평하다
```

```python
sns.kdeplot(df["Height"]) # Height에 대한 kde (꼭대기가 가파르다, 뾰족하다)
```

```python
sns.kdeplot(df["Weight"]) # Weight에 대한 kde (위쪽에는 제법 평평하다)
```

```python
df.plot(kind = "scatter", x = "Height", y = "Weight")
```

```python
sns.lmplot(data= df, x = "Height", y = "Weight") # Regression Line(회기선)이 그려짐 
# 키가 175인 학생은 몸무게가 대략 70kg일 것이다 라고 추측할 수 있다(데이터끼리 연관성이 높을수록 회기선도 정확하다)
```

```python
df = pd.read_csv("csv_file/laptops.csv")
df.head()
```

```python
df["os"]
```

```python
df["os"].unique() # 어떤 값들이 있는지 종류를 추려서 보여줌
```

```python
sns.catplot(data = df, x = "os", y = "price", kind = "box") # OS별로 가격데이터를 전반적으로 보고싶을 때
```

```python
sns.catplot(data = df, x = "os", y = "price", kind = "violin") # catplot -> category plot
```

```python
sns.catplot(data = df, x = "os", y = "price", kind = "strip") # kind = "strip" 데이터를 하나하나 볼 수 있다(디폴트 값임)
```

```python
df["processor_brand"].unique()
```

```python
# 어떤 값들이 인텔을 쓰고 어떤 값들이 amd를 쓰는지 알고 싶다
sns.catplot(data = df, x = "os", y = "price", kind = "strip", hue = "processor_brand") # hue는 색을 의미(프로세서 브랜드에 따라 색을 다르게)
```

```python
sns.catplot(data = df, x = "os", y = "price", kind = "swarm", hue = "processor_brand") 
# swarm -> strip이긴 한데, 데이터들이 겹쳐지지 않아서 보기 편하다
```
