# 카테고리로 분류

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv("csv_file/laptops.csv")
df.head()
```

```python
brand_nation = {
    "Dell" : "U.S",
    "Apple" : "U.S",
    "Dcer" : "Taiwan",
    "HP" : "U.S",
    "Lenovo" : "China",
    "Dell" : "U.S",
    "Alienware" : "U.S",
    "Asus" : "Taiwan"
}
```

```python
df["brand"].map(brand_nation) # brand_nation사전의 Key에 해당하는 값들을 모두 Value로 바꾸는 map 메서드
```

```python
df["brand_nation"] = df["brand"].map(brand_nation)
df
```

# groupby

```python
# groupby를 사용하면 특정 column을 기준으로 한 데이터 분류가 쉬워진다
```

```python
nation = df.groupby("brand_nation")
type(nation)
```

```python
nation.count() # 국가별로 각 column의 데이터 개수 출력
```

```python
nation.max() # 국가별로 각 column의 최대값 출력
```

```python
nation.mean() # 국가별로 각 column의 평균값 출력
```

```python
nation.first() # 국가별로 각 column의 첫번째값 출력
```

```python
nation.last() # 국가별로 각 column의 마지막값 출력
```

```python
nation.plot(kind = "box", y = "price") # groupby로 그래프도 그릴수 있음
```

```python
nation.plot(kind = "hist", y = "price") # groupby로 그래프도 그릴수 있음
```

# 데이터 합치기

### Merge 방법 4가지

1. inner join 
    - 두 데이터의 기준 column이 겹치는 것들에 대해서만 Merge
2. Left Outer join
    - 왼쪽 데이터의 기준 column 및 데이터  + 겹치는 것들 Merge
3. Left Outer join
    - 오른쪽 데이터의 기준 column 및 데이터  + 겹치는 것들 Merge
4. Full Outer join
    - 왼쪽 데이터의 기준 column 및 데이터 + 오른쪽 데이터의 기준 column 및 데이터 Mege

```python
price_df = pd.read_csv("csv_file/vegetable_price.csv")
quantity_df = pd.read_csv("csv_file/vegetable_quantity.csv")
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df, quantity_df, on = "Product", how = "inner")
```

```python
pd.merge(price_df, quantity_df, on = "Product", how = "left")
```

```python
pd.merge(price_df, quantity_df, on = "Product", how = "right")
```

```python
pd.merge(price_df, quantity_df, on = "Product", how = "outer")
```
