# *여행지 선정 과제*

```python
import pandas as pd
world_df = pd.read_csv("csv_file/world_cities.csv", index_col = 0)

world_df
```

```python
world_df["City / Urban area"].value_counts()
```

```python
world_df["Country"].value_counts()
```

```python
condition = ( world_df["Population"] / world_df["Land area (in sqKm)"]) > 10000
condition.value_counts()
```

```python
condition.describe()
```

```python
world_df["density"] =  world_df["Population"] / world_df["Land area (in sqKm)"]
world_df["density"].describe()
```

```python
world_df.sort_values(by="density", ascending = False)
```
