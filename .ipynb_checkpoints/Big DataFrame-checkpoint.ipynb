```python
df.sort_values(by="price") # 정렬 (자동 오름차순)
```

```python
df.sort_values(by="price", ascending = False) # 내림차순 정렬(ascending)
```

```python
df.sort_values(by="price", ascending = False, inplace = True) # 원본 데이터 변경(inplace)
df
```
